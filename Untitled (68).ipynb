```python
Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.
ans-Sure, here's how you can load the "fmri" dataset using the load_dataset function from the seaborn library, and create a line plot using timepoint and signal for different event and region columns:

python
Copy code
import seaborn as sns

# Load the fmri dataset
fmri = sns.load_dataset("fmri")

# Create the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", data=fmri)
In this code, sns.lineplot() creates a line plot using the timepoint column as the x-axis and the signal column as the y-axis. The hue parameter is used to color-code the lines based on the event column, and the style parameter is used to create different line styles based on the region column.

You can customize this plot further using other parameters available in sns.lineplot().





```

```python
Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.
ans-Sure, here's how you can use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column, with different colors for different values of the 'cut' column:

python
Copy code
import seaborn as sns

# Load the diamonds dataset
diamonds = sns.load_dataset("diamonds")

# Create the histogram with hue parameter for cut column
sns.histplot(data=diamonds, x="price", hue="cut", multiple="stack")
In this code, sns.histplot() creates a histogram using the price column as the x-axis. The hue parameter is used to color-code the bars based on the cut column, and the multiple parameter is used to stack the bars for each cut value.

You can customize this plot further using other parameters available in sns.histplot().





```

```python
Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
of the iris dataset.
ans-Sure! Here's the code to load the "iris" dataset from seaborn and plot a pair plot with hue parameter for the "species" column:

python
Copy code
import seaborn as sns

# Load the iris dataset
iris_data = sns.load_dataset('iris')

# Plot a pair plot with hue parameter
sns.pairplot(data=iris_data, hue='species')

# Show the plot
plt.show()
This code will create a pair plot of the "iris" dataset with the hue parameter set to the "species" column. The sns.pairplot() function creates a matrix of scatter plots for each pair of variables in the dataset, and the hue parameter specifies the column to use for color encoding. In this case, it will create different colors for each of the three species of iris flowers present in the dataset.

The resulting plot will show scatter plots for each pair of variables in the dataset, with each species of iris flowers represented by a different color. This plot can be useful for exploring the relationships between different variables in the dataset and for identifying patterns or trends in the data.





```

import seaborn as sns

# Load the flights dataset
flights = sns.load_dataset("flights")

# Pivot the dataset to create a matrix
flights_matrix = flights.pivot("month", "year", "passengers")

# Create the heatmap
sns.heatmap(data=flights_matrix, annot=True, fmt="d", cmap="YlGnBu")
In this code, flights.pivot() is used to pivot the dataset into a matrix where each row represents a month, each column represents a year, and the values represent the number of passengers. Then, sns.heatmap() creates a heatmap using the matrix. The annot parameter is used to show the values in each cell, the fmt parameter is used to format the values as integers, and the cmap parameter is used to set the color map.
